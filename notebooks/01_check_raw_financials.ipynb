```python
import pandas as pd

file_path = "../data/raw_financials.xlsx"

df = pd.read_excel(
    file_path,
    sheet_name="Raw_Financials",
    dtype={"stock_code": str}
)

df.columns = df.columns.str.strip()

df["stock_code"] = df["stock_code"].str.zfill(6)

df.head()
```

```python
import numpy as np

df_feature = df.copy()

# 按公司和年份排序，方便计算同比增长
df_feature = df_feature.sort_values(["stock_code", "year"]).reset_index(drop=True)

# 1. 统一境外收入口径：优先 exact，其次 proxy
df_feature["foreign_revenue_used"] = df_feature["exact_overseas_revenue"].fillna(
    df_feature["proxy_export_sales"]
)

# 2. 是否披露境外收入
df_feature["has_foreign_revenue_disclosure"] = df_feature["foreign_revenue_used"].notna().astype(int)

# 3. 境外收入占比
df_feature["foreign_revenue_ratio"] = df_feature["foreign_revenue_used"] / df_feature["revenue"]

# 4. 毛利与毛利率
df_feature["gross_profit"] = df_feature["revenue"] - df_feature["operating_cost"]
df_feature["gross_margin"] = df_feature["gross_profit"] / df_feature["revenue"]

# 5. 成本率
df_feature["cost_ratio"] = df_feature["operating_cost"] / df_feature["revenue"]

# 6. 归母净利率
df_feature["net_margin_attr"] = df_feature["net_profit_attr"] / df_feature["revenue"]

# 7. 经营现金流收入比
df_feature["ocf_margin"] = df_feature["operating_cash_flow"] / df_feature["revenue"]

# 8. 经营现金流 / 归母净利润
df_feature["ocf_to_net_profit"] = df_feature["operating_cash_flow"] / df_feature["net_profit_attr"]

# 9. 研发强度
df_feature["rnd_intensity"] = df_feature["rnd_expense"] / df_feature["revenue"]

# 10. 资本开支强度
df_feature["capex_intensity"] = df_feature["capex_cash_paid"] / df_feature["revenue"]

# 11. 平均存货与存货周转率
df_feature["avg_inventory"] = (
    df_feature["inventory_begin"] + df_feature["inventory_end"]
) / 2

df_feature["inventory_turnover"] = df_feature["operating_cost"] / df_feature["avg_inventory"]

# 12. 存货占收入比例
df_feature["inventory_to_revenue"] = df_feature["inventory_end"] / df_feature["revenue"]

# 13. 平均应收账款与应收账款周转率
df_feature["avg_ar"] = (
    df_feature["ar_begin"] + df_feature["ar_end"]
) / 2

df_feature["ar_turnover"] = df_feature["revenue"] / df_feature["avg_ar"]

# 14. 应收账款占收入比例
df_feature["ar_to_revenue"] = df_feature["ar_end"] / df_feature["revenue"]

# 15. 同比增长类特征
df_feature["revenue_growth_yoy"] = df_feature.groupby("stock_code")["revenue"].pct_change()
df_feature["net_profit_growth_yoy"] = df_feature.groupby("stock_code")["net_profit_attr"].pct_change()
df_feature["ocf_growth_yoy"] = df_feature.groupby("stock_code")["operating_cash_flow"].pct_change()
df_feature["inventory_growth_yoy"] = df_feature.groupby("stock_code")["inventory_end"].pct_change()
df_feature["ar_growth_yoy"] = df_feature.groupby("stock_code")["ar_end"].pct_change()

# 16. 清理无穷值
df_feature = df_feature.replace([np.inf, -np.inf], np.nan)

df_feature.head()
```
